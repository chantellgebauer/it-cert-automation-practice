```python
#Importamos el conjunto de datos de Flores Iris(Un clasico de la ciencia de datos)
from sklearn.datasets import load_iris

#Importamos el "cerebro" del algoritmo(el clasificador) y una utilidad
#para ver el arbol en texto
from sklearn.tree import DecisionTreeClassifier, export_text

#Importamos la herramienta para dividir nuestros datos en "grupo de estudio"
# y "examen final"
from sklearn.model_selection import train_test_split

#Importamos la libreria base de árboles para poder dibujarlos graficamente
from sklearn import tree

#Importamos la librería estándar para crear gráficos y ventanas de imagen
import  matplotlib.pyplot as plt

#cargamos el objeto que contiene las medidas de las flores y sus nombres
iris = load_iris()

#Guardamos en X las características (pétalos/sepalos) y en Y las etiquetas
#que flor es
X, y = iris.data, iris.target

#Dividimos los datos: el 80%(train) sirve para que el arbol aprenda y el 20%(test)
#para evaluar si aprendio

#'random_state=42' es una semilla para que los resultados sean siempre los mismos
#al ejecutar
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=102)

#Creamos el objeto del árbol. Ponemos un limite de 3 niveles de preguntas(max_depth) para
#para no complicarlo usamos 'gini' para que el arbol elija las preguntas que mejor separan
#las flores

clf = DecisionTreeClassifier(max_depth=5, criterion= 'gini')

#Aquí ocurre la magia: el algoritmo analiza los datos de entrenamiento
clf.fit(X_train, y_train)


#Usamos los datos del examen final (X_test) Para ver si el arbol adivina
score = clf.score(X_test, y_test)

#Mostramos el resultado final convertido a porcentaje
print(f"Precisión del modelo: {score * 100}%")


```

```python
#Definimos el tamaño de la ventana del gráfico (12 de ancho por 8 de alto )
plt.figure(figsize=(12,8))

#Esta es la función mágica que dibuja el diagrama del arbol de decisión
tree.plot_tree(clf,
               feature_names=iris.feature_names,#medidas
               class_names=iris.target_names,#tipo de flor
               filled=True) #el color de los cuadros indica especie y la intensidad
plt.show()
```
